```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_primers = pd.read_excel('./ysx008_supp_st_1.xlsx', names=['ID', 'Sequence', 'Keep'])
all_primers = all_primers[all_primers['Keep'] == 'Yes']
```

```python
all_primers.to_csv('./orthogonal_primers.csv')
```

```python
cut_sites= [
        ("BbsI", "GAAGAC"),
        ("BtgZI", "GCGATG"),
        ("BsaI", "GGTCTC"),
        ("BsmBI", "CGTCTC"),
        ("AarI", "CACCTGC"),
        ("BfuAI", "ACCTGC")]

def cutsite_check(seq, cut_sites=cut_sites):
    seq = str(seq)
    for enzyme in cut_sites:
        if sequence_search(enzyme[1],seq):
            return False
    return True
def sequence_search(search,sequence):
    if search in sequence or reverse_complement(search) in sequence:
        return True
    else:
        return False
def reverse_complement(seq):
    return seq.translate(str.maketrans("ATGC","TACG"))[::-1]
```

```python
for index, row in all_primers.iterrows():
    if not cutsite_check(row["Sequence"]):
        all_primers.drop([index], inplace=True)
        print(index)
```

```python
primers = all_primers[:96]['Sequence']

pairs = list()
for i in range(len(primers)):
    for j in range(i+1,len(primers)):
        pairs.append((primers.iloc[i], primers.iloc[j]))
        
len(pairs)
```

```python
pairs = pd.DataFrame(pairs, columns=["Forward", "Reverse"])
```

```python
pairs['GeneID'] = ""
```

```python
pairs.to_csv('./orthogonal_primers_pairs.csv')
```
