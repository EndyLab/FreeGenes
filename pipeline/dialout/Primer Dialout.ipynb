```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
all_primers = pd.read_excel('./ysx008_supp_st_1.xlsx', names=['ID', 'Sequence', 'Keep'])
all_primers = all_primers[all_primers['Keep'] == 'Yes']
```

```python
all_primers.to_csv('./orthogonal_primers.csv')
```

```python
primers = all_primers[:96]['Sequence']

pairs = list()
for i in range(len(primers)):
    for j in range(i+1,len(primers)):
        pairs.append((primers.iloc[i], primers.iloc[j]))
        
len(pairs)
```

```python
pairs = pd.DataFrame(pairs, columns=["Forward", "Reverse"])
```

```python
pairs['GeneID'] = ""
```

```python
pairs.to_csv('./orthogonal_primers_pairs.csv')
```
